```python
news_cl = news_1[~news_1.topic.isna()].copy()
```

```python
news_cl.head()
```

```python
news_cl.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
news_cl['topic_label'] = le.fit_transform(news_cl['topic'])
```

```python
news_cl.head()
```

```python
news_tr=news_cl.groupby(['topic','topic_label'])['title'].count().reset_index().sort_values(by='title',ascending=False).nlargest(4,'title')
news_tr
## Можно было сделать и так:news_tr=news_cl.groupby(['topic','topic_label'])['title'].count().reset_index().nlargest(4,'title')
```
